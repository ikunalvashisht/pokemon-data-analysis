# seaborn visualization

# countplot

```python
sns.countplot(x='Generation',data=df)
```

# barplot

```python
sns.barplot(x='Generation',y="Attack",data=df,hue="Legendary")
```

# violin plot

```python
sns.violinplot(x='Generation',y="Speed",data=df,hue="Legendary")
```

```python
sns.violinplot(x=df["Attack"])
```

# boxplot

```python
sns.boxplot(x="Generation",y="Defense",data=df,hue="Legendary")
```

```python
sns.boxplot(x=df["Attack"])
```

# distplot

```python
#take an empty canvas
plt.figure(figsize=(12,12))
plt.subplot(3,3,1)
sns.distplot(df.Attack)
plt.subplot(3,3,2)
sns.distplot(df.HP)
plt.subplot(3,3,3)
sns.distplot(df.Defense)
plt.subplot(3,3,4)
sns.distplot(df.Speed)
plt.subplot(3,3,5)
a=df["Sp. Atk"]
sns.distplot(a)
plt.subplot(3,3,6)
b=df["Sp. Def"]
sns.distplot(b)
```

# swarmplot

```python
sns.set_style("ticks")
sns.swarmplot(x="Generation",y='HP',data=df,hue="Legendary")
```

# lmplot

```python
sns.set_style("white")
sns.lmplot(x='Generation',y='HP',data=df,hue="Legendary")
```

# factorplot

```python
sns.set_style("dark")
sns.factorplot(x="Generation",y="Attack",data=df,hue="Legendary")
```

# jointplot

```python
sns.set_style("darkgrid")
sns.jointplot(x="Generation",y="Attack",data=df)
```

```python
sns.set_style("whitegrid")
sns.stripplot(x="Generation",y="Attack",data=df,hue="Legendary")
```

```python
sns.pairplot(df,hue="Legendary")
```
